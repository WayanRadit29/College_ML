```python
# Encoding dan Scaling 
df_processed = pd.get_dummies(df, columns=["gender"])
df_processed = df_processed.drop(columns=['id'])



# from sklearn.preprocessing import MinMaxScaler, StandardScaler

# num_col = df_processed.select_dtypes(include=("number")).columns
# scaler = MinMaxScaler()
# df_processed[num_col] = scaler.fit_transform(df_processed[num_col])













```

```python
# K-Means Clustering
from sklearn.cluster import KMeans

# Inisialisasi dan melatih model K-Means
kmeans = KMeans(
    n_clusters=2,      # Jumlah klaster yang ingin dibentuk
    init="k-means++",  # Metode inisialisasi centroid
    n_init=10,         # Berapa kali algoritma dicoba untuk hasil terbaik
    max_iter=300, 
             # Iterasi maksimum
)

kmeans.fit(df_processed)

# Menampilkan centroid hasil pelatihan
print(kmeans.cluster_centers_)

```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# 1 Ubah data jadi 2 dimensi
pca = PCA(n_components=2)
X_pca = pca.fit_transform(df_processed)

# 2 Ambil label cluster dan centroid-nya
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
centroids_pca = pca.transform(centroids)  # turunkan juga centroid ke 2D

# 3 Plot hasil clustering
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=labels, cmap="viridis", alpha=0.6)
plt.scatter(centroids_pca[:, 0], centroids_pca[:, 1], c="red", marker="X", s=200, label="Centroid")
plt.title("Visualisasi K-Means Clustering (PCA 2D)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.legend()
plt.show()
```

```python
pca = PCA(n_components=2)
X_train_2d = pca.fit_transform(df_processed)

plt.figure(figsize=(50, 40))
scatter = plt.scatter(X_train_2d[:, 0], X_train_2d[:, 1], c=kmeans.labels_, cmap='viridis', alpha=0.6)
plt.title('Visualisasi Clustering KMeans dengan PCA')
plt.xlabel('PC 1')
plt.ylabel('PC 2')
plt.colorbar(scatter, label='Cluster')
plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans

def compute_inertia_curve(X, k_min=1, k_max=10, random_state=42):
    K = list(range(k_min, k_max+1))
    inertia = []
    for k in K:
        km = KMeans(
            n_clusters=k,
            init="k-means++",
            n_init=10,          # atau 'auto' jika versi sklearn baru
            max_iter=300,
            random_state=random_state
        )
        km.fit(X)
        inertia.append(km.inertia_)
    return K, inertia

K, inertia = compute_inertia_curve(df_processed, k_min=1, k_max=10)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(7,5))
plt.plot(K, inertia, 'o-')
plt.xticks(K)
plt.xlabel("Jumlah cluster (k)")
plt.ylabel("Inertia (WCSS)")
plt.title("Elbow Method")
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
from sklearn.metrics import silhouette_score

def silhouette_scan(X, k_list, random_state=42):
    scores = {}
    for k in k_list:
        km = KMeans(n_clusters=k, init="k-means++", n_init=10, max_iter=300, random_state=random_state)
        labels = km.fit_predict(X)
        scores[k] = silhouette_score(X, labels)
    return scores

k_candidates = [2,3,4,5,6]
sil_scores = silhouette_scan(df_processed, k_candidates)
print(sil_scores)  # pilih k dengan skor tertinggi
```
